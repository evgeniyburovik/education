```python
student_performans = student_performans \
    .rename(columns =
            {'parental level of education': 'parental_levcl_of_education',
             'test preparation course': 'test_preparation_course',
             'math score': 'math_score',
             'reading score': 'reading_score',
             'writing score': 'writing_score'
})
```

```python
student_performans.head()
```

```python
student_performans.query('math_score > 88')
```

```python
student_performans.query("gender == 'female'")
```

```python
math_score_query = 74
```

```python
student_performans.query("math_score >= @math_score_query & gender == 'female'")
```

```python
list(student_performans)
```

```python
score_columns = [i for i in list(student_performans) if 'score' in i]
```

```python
student_performans[score_columns].head()
```

```python
student_performans.filter(like = 'score')
```

```python
student_performans.groupby('gender').count()
```

```python
student_performans.groupby('gender').mean()
```

```python
student_performans.groupby('race/ethnicity')\
    .aggregate({'math_score': 'mean', 'reading_score':'max'})
```

```python
student_performans.groupby('race/ethnicity', as_index = False)\
    .aggregate({'math_score': 'mean', 'reading_score':'max'}) \
    .rename(columns = {'math_score':'mean_math_score', 'reading_score':'max_reading_score'})
```

```python
student_performans.groupby(['race/ethnicity', 'gender'], as_index = False)\
    .aggregate({'math_score': 'mean', 'reading_score':'max'}) \
    .rename(columns = {'math_score':'mean_math_score', 'reading_score':'max_reading_score'})
```

```python
student_performans.groupby(['gender', 'race/ethnicity']).math_score.nunique()
```

```python
student_performans.sort_values(['gender','math_score'])
```

```python
student_performans.sort_values(['gender','math_score'], ascending = False) \
    .groupby('gender').head(5)
```

```python
student_performans['total_score'] = student_performans.math_score + student_performans.reading_score + student_performans.writing_score
```

```python
student_performans.head()
```

```python
student_performans = student_performans.assign(total_score_log = np.log(student_performans.total_score))
```

```python
 student_performans.head()
```

```python
 student_performans.drop(['total_score'], axis=1).head()
```
