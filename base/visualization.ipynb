```python
# Подгружаем библиотеки pandas и numpy
# numpy - Для работы с многомерными массивами и высокоуровневыми математическими функциями
# pandas - Работа pandas с данными строится поверх библиотеки NumPy, являющейся инструментом более низкого уровня 
# Предоставляет специальные структуры данных и операции для манипулирования числовыми таблицами и временны́ми рядами
import pandas as pd
import numpy as np
```

```python
# Подгружаем библиотеки matplotlib и seaborn
# Для работы с визуализацией данных
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns


# Подготавливаем общие настройки для графиков

sns.set(
    rc={
        "figure.figsize": (20, 10), #размер графика - 12 дюймов в ширину, 8 в длину
        "axes.titlesize": 20, #размер заголовка 
        "axes.labelsize": 14, #размер подписей осей
        "xtick.labelsize": 13, #размер подписей чёрточек на x-оси
        "ytick.labelsize": 13, #размер подписей чёрточек на y-оси
    }
)

sns.set_style("whitegrid") #тема фона, белая со штриховкой
sns.set(font_scale=2)
```

```python
# Закружаем датафрейм, странтартное разделение через запятую csv файл
student_performans = pd.read_csv('StudentsPerformance.csv')
```

```python
# Смотрим на структуру данных в табличной форме. Первые 10 значений по умолчанию
student_performans.head()
```

Переименуем столюцы для удобного обращения через метод

```python
student_performans = student_performans \
    .rename(columns = 
           {'parental level of education': 'parental_levcl_of_education',
             'test preparation course': 'test_preparation_course',
             'math score': 'math_score',
             'reading score': 'reading_score',
             'writing score': 'writing_score'
})
```

```python
student_performans.head()
```

```python
# Находим среднее
student_performans.math_score.mean()
```

```python
# Строим быструю гистограмму на основе бибилиотеки matplotlib
student_performans.math_score.hist()
```

```python
# Простое отобразение scatter диаграммы
student_performans.plot.scatter(x = 'math_score', y = 'reading_score')
```

```python
student_performans.plot.scatter(x = 'writing_score', y = 'reading_score')
```

```python
student_performans.plot.scatter(x = 'writing_score', y = 'math_score')
```

```python
sns.lmplot(x = 'math_score', y = 'reading_score', data = student_performans, hue="gender")
```

```python
sns.lmplot(x = 'math_score', y = 'reading_score', data = student_performans, hue="gender", fit_reg=False)
```

```python
ax = sns.lmplot(x = 'math_score', y = 'reading_score', data = student_performans, hue="gender", fit_reg=False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
