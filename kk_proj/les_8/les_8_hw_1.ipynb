```python
bikes_q2_sample_apr = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q2_sample_apr.csv')
bikes_q2_sample_jun = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q2_sample_jun.csv')
bikes_q2_sample_may = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q2_sample_may.csv')

bikes_q3_sample_aug = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q3_sample_aug.csv')
bikes_q3_sample_sep = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q3_sample_sep.csv')
bikes_q3_sample_july = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q3_sample_july.csv')

bikes_q4_sample_dec = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q4_sample_dec.csv')
bikes_q4_sample_nov = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q4_sample_nov.csv')
bikes_q4_sample_oct = pd.read_csv('/Users/evgeniyburovik/Documents/Courses/les_8/bikes_q4_sample_oct.csv')
```

```python
# Объединяем таблички эгноря индекс, чтобы от был с 0 до ...
bikes = pd.concat([
    bikes_q2_sample_apr ,bikes_q2_sample_jun, bikes_q2_sample_may,
    bikes_q3_sample_aug ,bikes_q3_sample_sep, bikes_q3_sample_july,
    bikes_q4_sample_dec ,bikes_q4_sample_nov, bikes_q4_sample_oct
                  ],ignore_index=True)
```

```python
bikes.head()
```

```python
bikes[['start_time', 'end_time']] = bikes[['start_time', 'end_time']].apply(pd.to_datetime)
```

```python
bikes.set_index('start_time', inplace=True)
```

```python
# Группируем по пользователям, бъем по дням и берем размер датафрейма(вернет количество строк, умноженное на количество столбцов)
group_usertype = bikes.groupby('usertype').resample('D').size().T
```

```python
bikes_summer = bikes.loc['2018-06-01':'2018-08-31']
```

```python
bikes_summer
```

```python
# Отбираем самый частый вариант станции и записываем в переменную с помощью функции idxmax
top_destination = bikes_summer.to_station_name.value_counts().idxmax()
```

```python
bad_day = bikes_summer.query('to_station_name == @top_destination')\
                        .resample('D')\
                        .size()\
                        .idxmin()\
                        .strftime('%Y-%m-%d')
```

```python
bad_day
```

```python
bad_day_1 = bikes_summer.query('to_station_name == @top_destination')\
                        .resample('D')\
                        .to_station_name.count()\
                        .idxmin()\
                        .strftime('%Y-%m-%d')
```

```python
bad_day_1
```

```python
bikes_summer['day_name'] = bikes_summer.end_time.dt.day_name()
```

```python
day_off = ['Saturday','Sunday']
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
```

```python
bikes_summer.groupby(['day_name', 'to_station_name'])\
            .agg({'trip_id':'count'})\
            .query('day_name in @day_off')\
            .sort_values('trip_id', ascending=False)
```

```python
bikes_summer.groupby(['day_name', 'to_station_name'], as_index=False)\
            .agg({'trip_id':'count'})\
            .query('day_name in @weekdays')\
            .sort_values('trip_id', ascending=False)
```
