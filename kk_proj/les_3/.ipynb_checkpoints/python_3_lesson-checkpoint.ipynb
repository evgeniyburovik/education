```python
user_id['brand_name'] = user_id.brand_info.apply(lambda x: x.split(' ')[-1])
```

```python
user_id
```

```python
user_purchases = user_id.groupby('user_id', as_index=False)\
                        .agg({'brand_name': 'count'})\
                        .rename(columns={'brand_name': 'count_purchases'})\
                        .query('count_purchases >= 5')
```

```python
lovely_brand_purchases_df = user_id.groupby(['user_id', 'brand_name'], as_index=False) \
        .agg({'brand_info':'count'}) \
        .sort_values(['user_id','brand_info'], ascending=[False, False]) \
        .groupby('user_id') \
        .head(1) \
        .rename(columns = {'brand_name':'lovely_brand','brand_info':'lovely_count'})
```

```python
user_unique_brands = user_id.groupby('user_id', as_index=False) \
        .agg({'brand_name': pd.Series.nunique}) \
        .rename(columns={'brand_name' : 'unique_brand'})
```

```python
loylity_df = user_purchases.merge(lovely_brand_purchases_df) \
              .merge(user_unique_brands)
```

```python
loylity_df
```

```python
loylity_df['loylity_score'] = loylity_df.lovely_count / loylity_df.count_purchases
```

```python
loylity_df
```

```python
plt.figure(figsize=(12, 8))
ax = sns.histplot(loylity_df.loylity_score)
```

```python
brand_loylity = loylity_df.groupby('lovely_brand', as_index=False) \
                            .agg({'loylity_score': 'median', 'user_id': 'count'}) \
                            .sort_values('loylity_score', ascending=False)
```

```python
brand_loylity
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='lovely_brand', y='loylity_score', data = brand_loylity)
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='lovely_brand', y='user_id', data = brand_loylity)
```
